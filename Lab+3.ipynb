# Accessing the ticker sybols & historical stock price of publicly lister companies

```python
import urllib.request
import pandas_datareader.data as web

def get_symbol(json_dict):
    #print (json_dict)
    resultList = json_dict['ResultSet']['Result']
    #print(len(resultList))
    if (len(resultList) > 0):
        #print(resultList)
        resultNAS = json_dict['ResultSet']['Result']
        for x in resultNAS:
            if (x['exch'] == 'NAS' or x['exch'] == 'NYSE'):
                return x['symbol']
    
        
def get_price(symbol):
    start = datetime.datetime(2004, 1, 1)
    end = datetime.datetime.today()
    f = web.DataReader(symbol, 'yahoo', start, end)
    df5 = f.to_frame()
    #print(df5)
    
def get_monthly_price(tickers):    
    data = web.get_data_yahoo(tickers,'01/01/2004','08/30/2017',interval='m')
    return data
        
    
my_list = df1.Entity.tolist()
tickers = []  
for xyz in my_list:
    url = "http://d.yimg.com/autoc.finance.yahoo.com/autoc?query=@&region=1&lang=en"
    url = url.replace("@",str(xyz))
    url = url.replace (" ","%20")
    
    try:
        # returns a utf-8 'bytes' object which still can't be processed
        json_data = urllib.request.urlopen(url).read()

        
        # decode into a string
        str_json_data = json_data.decode('utf-8')
        json_dict = json.loads(str_json_data)
    except:
        #print ('Cannot read JSON from URL:')
        #print (url)
        continue
    else:
        symbol = get_symbol(json_dict)
        if not (symbol is None):
            tickers.append(symbol)

print(len(tickers))
pricing_data_tickers = get_monthly_price(tickers)        

    

        
    
        
```
